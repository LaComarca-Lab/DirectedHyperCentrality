# Directed hypergraph statistics (PROCESSED)

```python
import pickle
import numpy as np
import networkx as nx
import xgi
import os
import matplotlib.pyplot as plt
from itertools import product, permutations
from collections import defaultdict

from useful_functions import hypergraph_statistics
```

## Chemical reactions

### Astrochem

#### 2F

```python
with open('ProcessedHypergraphs/astrochem2F.pkl', 'rb') as f:
    astrochem_data = pickle.load(f)
```

```python
H = xgi.DiHypergraph(astrochem_data)
#H.edges.dimembers()
```

IMPORTANT: Reactions such as H + H -> H2 require directed hyperedges with multiset collections (as in the Uplift paper with the multiply added auxiliary node). In XGI this is not contemplated.

```python
N, E, in_dict, out_dict, size_dict = hypergraph_statistics(astrochem_data)
```

```python
N, E, in_dict, out_dict, size_dict
```

```python
fig, ax  = plt.subplots(1,3, figsize=(15,5))

ax[0].bar(in_dict.keys(), in_dict.values())

ax[1].bar(out_dict.keys(), out_dict.values())

ax[2].bar([str(size) for size in size_dict.keys()], size_dict.values())

ax[2].set_xticks(ax[2].get_xticks())
ax[2].set_xticklabels(ax[2].get_xticklabels(), rotation=90, ha='right')

plt.show()
```
