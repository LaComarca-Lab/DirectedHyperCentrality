## Metabolical reactions

### BiGG

```python
NEdict = {}

folder = 'ProcessedHypergraphs/Metabolical-BiGG/'
for file in os.listdir(folder):

    with open(folder + file, 'rb') as f:
        BiGG_data = pickle.load(f)
        
    N, E, _, _, _ = hypergraph_statistics(BiGG_data)
    
    NEdict[file.split('.pkl')[0]] = (N,E)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,4))

n1, n2, n3 = (0,0,0)
for key, (N,E) in NEdict.items():
    
    
    if '2F' in key:
        if N > 39:
            ax1.scatter(N,E, label=key)
            
            n1 += 1
        else:
            ax1.scatter(N,E, label=None)
            ax2.scatter(N,E, label=None)
            
            n1 += 1
            n2 += 1
        
    elif '3F' in key:
        ax3.scatter(N,E, label=key)
        
        n3 += 1
        
fig.supylabel('E')
fig.supxlabel('N')    

ax1.legend()
ax3.legend()

ax1.set_title(f"2F ({n1} hypergraphs)")
ax2.set_title(f"Zoomed 2F ({n2} hypergraphs)")
ax3.set_title(f"3F ({n3} hypergraphs)")

plt.tight_layout()
```
